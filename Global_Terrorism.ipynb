**Project: Data Processing and Cleaning on the Global Terrorism Database**

Objective:

The goal of this project is to preprocess and clean the Global Terrorism Database to prepare it for further analysis. This project will involve handling missing values, dealing with inconsistencies, and preparing the dataset for various analytical tasks.

Dataset:

**Name**: Global Terrorism Database

**Source:** Global Terrorism Database on Kaggle [link](https://www.kaggle.com/datasets/START-UMD/gtd?resource=download)

This dataset contains over 180,000 terrorist incidents globally from 1970 to the present, with information on the date, location, attackers, and types of attacks.


**2.HANDLING MISSING DATA**

```python
data.isnull().sum()  # Count missing values in each column
```

```python
# Dropping irrelevant columns (example: columns with >70% missing values)
columns_to_drop = data.columns[data.isnull().mean() > 0.7]
data_cleaned = data.drop(columns=columns_to_drop)
data_cleaned
```

```python
# Filling missing values with mean for numerical columns and mode for categorical columns
data_cleaned['nkill'].fillna(data_cleaned['nkill'].mean(), inplace=True)
data_cleaned['nwound'].fillna(data_cleaned['nwound'].mean(), inplace=True)
```

**3. Handling Duplicates**

```python
data_cleaned = data_cleaned.drop_duplicates()
```

**4. Handling Outliersext**

```python
# Example: Detecting outliers in 'nkill' column using Interquartile Range (IQR)
Q1 = data_cleaned['nkill'].quantile(0.25)
Q3 = data_cleaned['nkill'].quantile(0.75)
IQR = Q3 - Q1

# Define limits to detect outliers
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

# Identifying and capping outliers
data_cleaned['nkill'] = data_cleaned['nkill'].clip(lower=lower_limit, upper=upper_limit)
```

**5. Converting Data Types**

```python

# Convert categorical data types (e.g., attack type, weapon type) to category
data_cleaned['weaptype1_txt'] = data_cleaned['weaptype1_txt'].astype('category')
data_cleaned['attacktype1_txt'] = data_cleaned['attacktype1_txt'].astype('category')
```

```python
# One-hot encoding for categorical columns like 'attacktype1_txt'
data_cleaned = pd.get_dummies(data_cleaned, columns=['attacktype1_txt', 'weaptype1_txt'], drop_first=True)
```

**6.Scaling and Normalization**

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Using StandardScaler to standardize numerical columns
scaler = StandardScaler()
data_cleaned[['nkill', 'nwound']] = scaler.fit_transform(data_cleaned[['nkill', 'nwound']])
```

**7.Data Visualization**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Visualizing the number of attacks by year
sns.countplot(x='country_txt', data=data_cleaned)
plt.title('Number of Terrorist Attacks per country')
plt.show()
```

```python
# Visualizing the number of attacks by year
sns.countplot(x='iyear', data=data_cleaned)
plt.title('Number of Terrorist Attacks per Year')
plt.show()
```

```python
# Group data by year and count the number of attacks each year
attacks_per_year = data_cleaned.groupby('iyear').size()

# Plot the number of attacks per year
plt.figure(figsize=(10, 6))
sns.lineplot(x=attacks_per_year.index, y=attacks_per_year.values)
plt.title('Number of Terrorist Attacks Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.grid(True)
plt.show()
```

```python
# Count the number of attacks per country
top_countries = data_cleaned['country_txt'].value_counts().nlargest(10)

# Plot the top 10 countries with the most attacks
plt.figure(figsize=(10, 6))
sns.barplot(x=top_countries.values, y=top_countries.index, palette="viridis")
plt.title('Top 10 Countries with the Most Terrorist Attacks')
plt.xlabel('Number of Attacks')
plt.ylabel('Country')
plt.show()
```

```python
# Group data by year and region and count the number of attacks
attacks_per_year_region = data_cleaned.pivot_table(index='region_txt', columns='iyear', aggfunc='size', fill_value=0)

# Create a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(attacks_per_year_region, cmap="YlGnBu", linewidths=0.5)
plt.title('Heatmap of Terrorist Attacks by Year and Region')
plt.xlabel('Year')
plt.ylabel('Region')
plt.show()
```
